```python
# --- SIMULADOR INTEGRADO DE IA PARA LABORATORIO ---

# El programa inicia solicitando la identificación del usuario [1]
print("INICIALIZANDO SISTEMA DE IA: 'LAB-BOT 2026'...")
nombre_cientifico = input("Identifíquese, Investigador: ")
print(f"Bienvenido, Dr/Dra. {nombre_cientifico}. Acceso concedido.")
print() # Espaciado solicitado

# --- SECCIÓN 1: ANÁLISIS DE SEGURIDAD TÉRMICA (Operadores Relacionales) ---
# Se utiliza la entrada para evaluar límites de seguridad [1, 2]
temperatura = float(input("Ingrese la temperatura actual del reactor (°C): "))

if temperatura > 120: # Operador relacional '>' [3]
    print("ESTADO: ¡ALERTA CRÍTICA! Temperatura excedida. Active sistema de enfriamiento.")
elif temperatura >= 0 and temperatura <= 120: # Operadores relacionales y lógico 'and' [3, 4]
    print("ESTADO: Operación normal. Temperatura dentro de los parámetros de diseño.")
else:
    print("ESTADO: ¡AVISO! Temperatura bajo cero. Verifique estado de la muestra criogénica.")
print() # Espaciado solicitado

# --- SECCIÓN 2: CLASIFICACIÓN QUÍMICA (Operadores Aritméticos y Relacionales) ---
# Se evalúa el pH utilizando comparaciones específicas [5, 6]
ph_muestra = float(input("Ingrese el nivel de pH detectado en la muestra: "))

if ph_muestra == 7.0: # Operador de igualdad [3]
    print("CLASIFICACIÓN IA: La sustancia es Neutra (Agua destilada o similar).")
elif ph_muestra > 7.0 and ph_muestra <= 14.0:
    print("CLASIFICACIÓN IA: La sustancia es una Base (Alcalina).")
elif ph_muestra >= 0.0 and ph_muestra < 7.0:
    print("CLASIFICACIÓN IA: La sustancia es un Ácido.")
else:
    print("ERROR: El valor de pH ingresado está fuera de la escala química estándar (0-14).")
print() # Espaciado solicitado

# --- SECCIÓN 3: CÁLCULO DE RENDIMIENTO (Operadores Aritméticos Complejos) ---
# Cálculo de masa molar teórica simulada [6]
masa_reactivo = float(input("Ingrese la masa del reactivo utilizado (gramos): "))
volumen_solvente = float(input("Ingrese el volumen del solvente (litros): "))

# Operador aritmético de división '/' y potencia '**' [6]
concentracion_final = masa_reactivo / volumen_solvente
eficiencia_proceso = (masa_reactivo ** 0.8) # Potencia simulada para el modelo de IA

print(f"RESULTADOS CALCULADOS: Concentración de {concentracion_final:.2f} g/L.")
print(f"ESTIMACIÓN IA: El índice de eficiencia estimado es {eficiencia_proceso:.2f}.")
print() # Espaciado solicitado

# --- SECCIÓN 4: DECISIÓN DE SÍNTESIS (Combinación de Operadores Lógicos) ---
# La IA decide si procede la síntesis basada en múltiples factores [3, 7]
print("EVALUANDO VIABILIDAD DE SÍNTESIS FINAL...")

# Verificación compleja usando 'or', 'and' y 'not' [3]
if (concentracion_final > 10.0 and ph_muestra == 7.0) or (temperatura > 20.0 and temperatura < 25.0):
    print("PREDICCIÓN IA: Condiciones de síntesis ÓPTIMAS. Puede proceder con el experimento.")
else:
    print("PREDICCIÓN IA: Riesgo de inestabilidad detectado. Ajuste los parámetros y reintente.")
print() # Espaciado solicitado

print(f"SISTEMA IA LAB-BOT: Sesión cerrada para {nombre_cientifico}. Reporte generado.")
print("Programa realizado por Alonso Lopez Garcia")
```
